```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Generate histogram
plt.hist(strengths, bins=40, alpha=0.5, density=True, edgecolor='black')

# Get mean and standard deviation of the data
mu, std = norm.fit(strengths)

# Generate distribution curve
x = np.linspace(0, max(strengths), 100)
p = norm.pdf(x, mu, std)

# Plot distribution curve
plt.plot(x, p, 'k', linewidth=1)

# Set plot labels
plt.title('Distribution of Password Strengths Using a First-Order Markov Model')
plt.xlabel('Password Strength')
plt.ylabel('Probabilistic Density')

# Show plot
plt.show()
```
